## DataFrame

DataFrame é uma estrutura de dados bidimensional, ou seja, os dados são alinhados de maneira tabular em linhas e colunas.

***pandas.DataFrame( data, index, columns, dtype, copy)***
1.   ***data:*** Dados assumem várias formas como ndarray, series, maps, lists, dict, constants e também outros DataFrames.
2.   ***index:*** Para rotular linhas, o índice a ser usado para o DataFrame resultante é opcional. Default np.arrange(n) se nenhum índice for passado.
3. ***columns:*** Para rotular colunas, a sintaxe padrão opcional é: np.arrange (n). Isso só é verdade se nenhum índice for passado.
4. ***dtype:*** Tipo de dados de cada coluna.
5. ***copy:*** Este comando (ou seja o que for) é usado para copiar dados, se o padrão for Falso.

# K-nearest Neighbours

É um algoritmo simples e poderoso de aprendizado supervisionado usado para classificação e regressão. Ele funciona encontrando os k vizinhos mais próximos de um ponto novo (baseado em uma métrica de distância, como a euclidiana) e determinando sua categoria ou valor com base nesses vizinhos.

***Exemplo prático:*** Imagine que você quer classificar uma fruta como maçã ou laranja. Você tem um gráfico onde cada ponto representa frutas conhecidas, com características como peso e cor. Se você inserir uma nova fruta, o KNN vai olhar para as frutas mais próximas no gráfico (os vizinhos) e decidir com base na maioria. Se 3 dos 5 vizinhos forem maçãs, a nova fruta será classificada como maçã.

```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
irs = pd.DataFrame(iris.data, columns=iris.feature_names)
irs['class'] = iris.target
irs.head()

x = irs.iloc[:, :-1].values
y = irs.iloc[:, 4].values
```

## Divisão em conjunto de teste e de treino

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

### Normalização

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

## Treinamentos e previsões

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
```

### Avaliando o algoritmo

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
